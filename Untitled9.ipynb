```python
#Data Processing
print(df.info())  # View column types and missing values
print(df.describe())  # Summary statistics
print(df.head())  # Preview the first few rows
```

```python
df = df.dropna(subset=['Class'])  # Remove rows where 'Class' is NaN
X = df.drop(columns=['Class'])    # Features
y = df['Class']                   # Target
```

```python
# Separate features and target
X = df.drop(columns=['Class'])  # Features
y = df['Class']                 # Target (Fraud or Non-Fraud)

from imblearn.over_sampling import SMOTE
from collections import Counter

# Initialize SMOTE with fewer neighbors
smote = SMOTE(random_state=42, k_neighbors=1)

# Apply SMOTE to balance the dataset
X_resampled, y_resampled = smote.fit_resample(X, y)

# Verify the new class distribution
print("Class distribution after SMOTE:", Counter(y_resampled))
```

```python
#Plotting Class Distribution After SMOTE

from collections import Counter

# Get the class distribution
class_counts = Counter(y_resampled)

# Plot the class distribution
plt.bar(class_counts.keys(), class_counts.values(), color=['blue', 'orange'])
plt.title('Class Distribution After SMOTE')
plt.xlabel('Class')
plt.ylabel('Number of Samples')
plt.xticks([0, 1], ['Non-Fraud (0)', 'Fraud (1)'])  # Replace with your class labels if different
plt.show()
```

```python
# Feature Scaling
from sklearn.preprocessing import StandardScaler

# Initialize the scaler
scaler = StandardScaler()

# Apply scaling
X_scaled = scaler.fit_transform(X_resampled)

# Convert scaled features back to a DataFrame (optional, for clarity)
import pandas as pd
X_scaled_df = pd.DataFrame(X_scaled, columns=X.columns)
```

```python
# Before scaling
X_resampled_df = pd.DataFrame(X_resampled, columns=X.columns)

# Plot before and after scaling
fig, axes = plt.subplots(1, 2, figsize=(15, 6))

# Original data distribution
X_resampled_df.boxplot(ax=axes[0], vert=False, showfliers=False)
axes[0].set_title('Feature Distribution Before Scaling')
axes[0].set_xlabel('Values')
axes[0].set_ylabel('Features')

# Scaled data distribution
X_scaled_df.boxplot(ax=axes[1], vert=False, showfliers=False)
axes[1].set_title('Feature Distribution After Scaling')
axes[1].set_xlabel('Values')

plt.tight_layout()
plt.show()
```

```python
#Data Splitting
from sklearn.model_selection import train_test_split

# Split the data
X_train, X_test, y_train, y_test = train_test_split(
    X_scaled, y_resampled, test_size=0.2, random_state=42
)

# Print shapes of the splits
print("Training set shape:", X_train.shape)
print("Test set shape:", X_test.shape)
```

```python
import matplotlib.pyplot as plt

# Sizes of splits
split_sizes = [len(y_train), len(y_test)]
labels = ['Training Data', 'Testing Data']

# Plot pie chart
plt.figure(figsize=(6, 6))
plt.pie(split_sizes, labels=labels, autopct='%1.1f%%', startangle=90, colors=['skyblue', 'orange'])
plt.title('Train-Test Split Ratio')
plt.show()
```

```python
#heatmap
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix
corr_matrix = df.corr()

# Plot the heatmap for all features
plt.figure(figsize=(12, 8))
sns.heatmap(corr_matrix, annot=False, cmap='coolwarm', cbar=True)
plt.title('Correlation Heatmap Between Features')
plt.show()

# Find correlations with 'Class' column
correlations = corr_matrix['Class'].sort_values(ascending=False)
print("Top Correlated Features with Class:\n", correlations)
```

```python
#Correlation Matrix
#What we do: Find the features most correlated with the target variable (Class).
import seaborn as sns
import matplotlib.pyplot as plt

# Compute the correlation matrix with target variable 'Class'
correlation = df.corrwith(df['Class']).sort_values(ascending=False)
print("Features Correlated with 'Class':\n", correlation)

# Plot correlation with target
plt.figure(figsize=(8, 6))
correlation.drop('Class').plot(kind='bar', color='skyblue')
plt.title("Feature Correlation with Target (Class)")
plt.ylabel("Correlation")
plt.xlabel("Features")
plt.show()

# Select features with correlation > 0.1 (absolute value)
selected_features_corr = correlation[abs(correlation) > 0.1].index.tolist()
print("Features Selected After Correlation:", selected_features_corr)

# Drop target variable 'Class' from selected features
selected_features_corr.remove('Class')
```

```python
#Recursive Feature Elimination (RFE)
#What we do: Refine the features further using RFE to eliminate less important features.
from sklearn.feature_selection import RFE
from sklearn.linear_model import LogisticRegression

# Initialize Logistic Regression model
log_reg = LogisticRegression(max_iter=1000, random_state=42)

# Initialize RFE to select top 10 features
rfe = RFE(estimator=log_reg, n_features_to_select=10)

# Apply RFE on features selected from correlation
X_corr = df[selected_features_corr]
y = df['Class']
rfe.fit(X_corr, y)

# Get selected features
selected_features_rfe = X_corr.columns[rfe.support_]
print("Features Selected After RFE:", selected_features_rfe)

# Update dataset with RFE features
X_rfe = X_corr[selected_features_rfe]


```

```python
#Feature Importance Using Random Forest
#What we do: Use Random Forest to rank feature importance.
from sklearn.ensemble import RandomForestClassifier
import pandas as pd

# Train Random Forest on RFE features
rf = RandomForestClassifier(random_state=42)
rf.fit(X_rfe, y)

# Get feature importance
feature_importance = pd.Series(rf.feature_importances_, index=X_rfe.columns).sort_values(ascending=False)
print("Feature Importance:\n", feature_importance)

# Plot feature importance
plt.figure(figsize=(8, 6))
feature_importance.plot(kind='bar', color='lightgreen')
plt.title("Feature Importance from Random Forest")
plt.ylabel("Importance Score")
plt.xlabel("Features")
plt.show()
```

```python
# Model Selection
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, roc_auc_score, confusion_matrix
from sklearn.linear_model import LogisticRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.ensemble import RandomForestClassifier
from xgboost import XGBClassifier
from sklearn.svm import SVC
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# Train and Evaluate Models
#Logistic Regression
# Train Logistic Regression
log_reg = LogisticRegression(max_iter=1000, random_state=42)
log_reg.fit(X_train, y_train)

# Predict and Evaluate
y_pred = log_reg.predict(X_test)
y_proba = log_reg.predict_proba(X_test)[:, 1]

print("Logistic Regression Report:")
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
#Decision Tree
# Train Decision Tree
dt_model = DecisionTreeClassifier(random_state=42)
dt_model.fit(X_train, y_train)

# Predict and Evaluate
y_pred = dt_model.predict(X_test)
y_proba = dt_model.predict_proba(X_test)[:, 1]

print("Decision Tree Report:")
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
# Train Random Forest
rf_model = RandomForestClassifier(n_estimators=100, random_state=42)
rf_model.fit(X_train, y_train)

# Predict and Evaluate
y_pred = rf_model.predict(X_test)
y_proba = rf_model.predict_proba(X_test)[:, 1]

print("Random Forest Report:")
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
# Train XGBoost Classifier
xgb_model = XGBClassifier(use_label_encoder=False, eval_metric='logloss', random_state=42)
xgb_model.fit(X_train, y_train)

# Predict and Evaluate
y_pred = xgb_model.predict(X_test)
y_proba = xgb_model.predict_proba(X_test)[:, 1]

print("XGBoost Report:")
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
# Train Support Vector Machine
svm_model = SVC(probability=True, random_state=42)
svm_model.fit(X_train, y_train)

# Predict and Evaluate
y_pred = svm_model.predict(X_test)
y_proba = svm_model.predict_proba(X_test)[:, 1]

print("Support Vector Machine Report:")
print(classification_report(y_test, y_pred))
print("ROC-AUC Score:", roc_auc_score(y_test, y_proba))
```

```python
#ROC-AUC Scores: Shows the model's ability to distinguish between fraud and non-fraud.
# Collect ROC-AUC scores
model_scores = {
    "Logistic Regression": roc_auc_score(y_test, log_reg.predict_proba(X_test)[:, 1]),
    "Decision Tree": roc_auc_score(y_test, dt_model.predict_proba(X_test)[:, 1]),
    "Random Forest": roc_auc_score(y_test, rf_model.predict_proba(X_test)[:, 1]),
    "XGBoost": roc_auc_score(y_test, xgb_model.predict_proba(X_test)[:, 1]),
    "SVM": roc_auc_score(y_test, svm_model.predict_proba(X_test)[:, 1])
}

# Plot ROC-AUC scores
plt.figure(figsize=(8, 6))
plt.bar(model_scores.keys(), model_scores.values(), color='skyblue')
plt.title("Model Comparison: ROC-AUC Scores")
plt.ylabel("ROC-AUC Score")
plt.xticks(rotation=45)
plt.show()

# Print scores
for model, score in model_scores.items():
    print(f"{model}: {score:.4f}")
```

```python
#Model Evaluation
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score, roc_auc_score, classification_report

# Create a dictionary to store evaluation results
evaluation_metrics = {}

# Helper function to calculate metrics for a model
def evaluate_model(name, model, X_test, y_test):
    y_pred = model.predict(X_test)
    y_proba = model.predict_proba(X_test)[:, 1]  # Probabilities for ROC-AUC

    metrics = {
        "Accuracy": accuracy_score(y_test, y_pred),
        "Precision": precision_score(y_test, y_pred),
        "Recall": recall_score(y_test, y_pred),
        "F1-Score": f1_score(y_test, y_pred),
        "ROC-AUC": roc_auc_score(y_test, y_proba)
    }

    print(f"--- {name} ---")
    for metric, value in metrics.items():
        print(f"{metric}: {value:.4f}")
    print("\n")

    return metrics

# Evaluate all models
evaluation_metrics["Logistic Regression"] = evaluate_model("Logistic Regression", log_reg, X_test, y_test)
evaluation_metrics["Decision Tree"] = evaluate_model("Decision Tree", dt_model, X_test, y_test)
evaluation_metrics["Random Forest"] = evaluate_model("Random Forest", rf_model, X_test, y_test)
evaluation_metrics["XGBoost"] = evaluate_model("XGBoost", xgb_model, X_test, y_test)
evaluation_metrics["SVM"] = evaluate_model("SVM", svm_model, X_test, y_test)
```

```python
# Convert evaluation metrics to a DataFrame
metrics_df = pd.DataFrame(evaluation_metrics).T  # Transpose for better readability

# Print the DataFrame
print("Model Evaluation Summary:")
print(metrics_df)

# Plot the metrics for comparison
metrics_df.plot(kind="bar", figsize=(10, 6))
plt.title("Model Evaluation Metrics Comparison")
plt.ylabel("Scores")
plt.xlabel("Models")
plt.xticks(rotation=45)
plt.legend(loc="lower right")
plt.show()
```

```python
#Bayesian Optimization
from bayes_opt import BayesianOptimization
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import cross_val_score

# Define the function to optimize
def optimize_log_reg(C, penalty):
    penalty = 'l1' if penalty < 0.5 else 'l2'  # Convert numeric value to 'l1' or 'l2'
    model = LogisticRegression(C=C, penalty=penalty, solver='liblinear')  # Create model with given parameters
    return cross_val_score(model, X_train, y_train, cv=5, scoring='roc_auc').mean()  # 5-fold cross-validation

# Define the parameter bounds
pbounds = {'C': (0.1, 10.0), 'penalty': (0, 1)}

# Perform Bayesian Optimization
optimizer = BayesianOptimization(f=optimize_log_reg, pbounds=pbounds, random_state=42)
optimizer.maximize(init_points=2, n_iter=100)  # Initial points for exploration, total iterations

# Best parameters and score
best_params = optimizer.max['params']
best_params['penalty'] = 'l1' if best_params['penalty'] < 0.5 else 'l2'  # Convert numeric value to 'l1' or 'l2'
print(f"Best parameters: {best_params}")  # Print the best parameters
print(f"Best ROC-AUC Score: {optimizer.max['target']:.4f}")  # Print the best ROC-AUC score
```
